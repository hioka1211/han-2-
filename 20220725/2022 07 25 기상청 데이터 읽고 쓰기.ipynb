```python
#모든 해의 8월달 데이터를 리스트로 담아 시각화
import csv

with open('./seoul.csv', 'r') as f:
    data = csv.reader(f)
    header = next(data)
    month_8=[]
    for row in data:
        try:
            #row[0]은 전체 날짜, 월과 날짜인 [5:7], append로 8월 추가
            if '08'==row[0][5:7]:
                month_8.append(float(row[-1]))
        except:
            pass
        
print(month_8)        
```

```python
#문제1 본인 생일의 최고 기온을 그래프로 그리기
```

```python
import csv
f = open('seoul.csv', 'r')
data = csv.reader(f)
header = next(data)
result = []
max_tmp = []
for row in data:
    if row[-1] != '': #최고 기온 값이 존재한다면
        if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '11': #12월11일에 해당하는 값이라면
            result.append(float(row[-1])) #리스트에 최고 기온 값 추가
        
plt.plot(result, 'red') #그래프에 색 추가
plt.show

f.close()
```

```python
import csv
f = open('seoul.csv', 'r')
data = csv.reader(f)
header = next(data)
result = []
max_tmp = [] #최고 기온 값
min_tmp = [] #최저 기온 값
for row in data:
    if row[-1] != '' and row[-2] != '': #최고 기온 값과 최저 기온 값이 존재한다면
        if row[0][5:] == '12-11':        #12월 11일에 존재하는 값
            max_tmp.append(float(row[-1]))
            min_tmp.append(float(row[-2]))
plt.plot(max_tmp)
plt.plot(min_tmp)
plt.show

f.close()
```

```python
#문제 3 2015년 데이터만 추출하여 저장하기
```

```python
import csv

read_f = open('./seoul.csv','r')
write_f = open('2015.csv', 'w',newline='')

data_r = csv.reader(read_f)
data_w = csv.writer(write_f)

#header(컬러명)부터 기록
header=next(data_r)
data_w.writerow(header)              
              
#data_r의 자료를 한줄씩 읽어서 2015년이면
#data_w에 기록한다

for row in data_r:
    if row[0][:4] == '2015':
        data_w.writerow(row)
    
write_f.close()              
read_f.close()
```
