```python
#생존자, 사망자 평균 나이 구하기
train_data.head()
```

```python
사망자=round(train_data.groupby('Survived')['Age'].mean()[0])
생존자=round(train_data.groupby('Survived')['Age'].mean()[1])
```

```python
print(사망자,생존자)
```

```python
train_data.head()
```

```python
사망자표=train_data[train_data['Survived']==0]
```

```python
생존자표=train_data[train_data['Survived']==1]
```

```python
사망자표['Age']=사망자표['Age'].fillna(사망자)
```

```python
사망자표.isnull().sum()

생존자표['Age']=생존자표['Age'].fillna(생존자)

생존자표.isnull().sum()

res_df=pd.concat([사망자표,생존자표])

res_df=res_df.reset_index(drop=True)

train_data.head()

train_data['Age']=train_data[train_data['Survived']==0]['Age'].fillna(사망자)

train_data['Age']=train_data[train_data['Survived']==1]['Age'].fillna(생존자)
```
