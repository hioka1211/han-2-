```python
from PIL import Image
import os
import pandas as pd
path='./images/'
files=os.listdir(path)

res=[]
for file in files:
    temp=path+file
    
    img = Image.open(temp)
    res.append([img.filename,img.format,img.size,img.mode,img.width,img.height])
df=pd.DataFrame(res,columns=['파일명','형식','크기','모드','가로크기','세로크기'])
#print(type(df))
display(df)
df.to_csv('이미지정보.csv',encoding='cp949')
```
