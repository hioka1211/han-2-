```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome('./chromedriver.exe')

date=['20220807','20220808']

url='https://www.melon.com/chart/day/index.htm'
driver.get(url)
time.sleep(2)
html=driver.page_source
# driver.close()
# driver.quit()
```

```python
type(html)
```

```python
soup=bs(html,'html.parser')
```

```python
div_tags=soup.find('div','service_list_song type02 d_song_list')
tr_tags=div_tags.find_all('tr')
```

```python
pd.date_range('20220807')
```

```python
res=[]
for tr in tr_tags[1:]:
    rank=tr.find('span','rank').text
    song=tr.find('div','wrap_song_info').find('a').text
    art=tr.find('div','ellipsis rank02').find('a').text
    alb=tr.find('div','ellipsis rank03').find('a').text
    tmp={
        '날짜':
        '순위':rank,
        '곡명':song,
        '아티스트':art,
        '앨범명':alb
    }
    res.append(tmp)
df=pd.DataFrame(res)
display(df)

df['날짜']='20220808'
```
