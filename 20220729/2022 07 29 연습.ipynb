```python
ls=['period','expense','overal']
for i in ls:
    plt.figure(figsize=(15,6))
    plt.subplot(1,2,1)
    plt.scatter(df.index, df[i])
    plt.title(f'{i}산점도')
    plt.subplot(1,2,2)
    plt.boxplot(df[i])
    plt.title(f'{i}박스플롯')
    plt.savefig(f'{i}.png')
    plt.show()
```

```python
#국가별 체류기간 합을 차트화
a=df.groupby('nat')['period'].sum()
a.plot(kind='bar')
plt.title('국가별 체류기간 합')
```

```python
ls=['period','expense','overal']
df_nat = df.groupby('nat')
plt.figure(figsize=(15,6))
for idx,i in enumerate(ls):
    plt.subplot(1,3,idx+1)
    df_nat[i].mean().sort_values().plot(kind='barh')
    plt.title(f'국가별 {i} 평균')

plt.tight_layout()
plt.show()
```

```python
nats=df['nat'].unique()
ls=['period','expense','overal']
for nat in nats:
    temp=df[df['nat']==nat]
    plt.figure(figsize=(15,6)) #차트 세팅
    for idx,i in enumerate(ls):
        plt.subplot(1,len(ls),idx+1)
        plt.scatter(temp.index,temp[i])
        plt.title(nat+i)

```
