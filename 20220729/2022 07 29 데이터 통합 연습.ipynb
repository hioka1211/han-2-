```python
CCTV_Seoul.rename(columns={CCTV_Seoul.columns[0]:'구별'}, inplace=True)
CCTV_Seoul.head()                  
```

```python
pop_Seoul = pd.read_excel('../data/01. population_in_Seoul.xls', header = 2, usecols='B,D,G,J,N')
pop_Seoul.head()
```

```python
pop_Seoul.columns=['구별','인구수','한국인','외국인','고령자']
```

```python
pop_Seoul.head()
```

```python
CCTV_Seoul['최근증가율']=CCTV_Seoul.iloc[:,-3:].sum(axis=1)/CCTV_Seoul['소계']*100
```

```python
CCTV_Seoul.head()
```

```python
pop_Seoul=pop_Seoul[1:]
```

```python
pop_Seoul=pop_Seoul.reset_index(drop=True)
```

```python
pop_Seoul.head()
```

```python
pop_Seoul['외국인비율']=pop_Seoul['외국인']/pop_Seoul['인구수']*100
```

```python
pop_Seoul['고령자비율']=pop_Seoul['고령자']/pop_Seoul['인구수']*100
```

```python
pop_Seoul.head()
```

```python
CCTV_Seoul.head()
```

```python
df_result=pd.merge(CCTV_Seoul,pop_Seoul,on='구별')
```

```python
df_result.columns
```

```python
df_result=df_result[['구별', '소계', '최근증가율', '인구수',
       '한국인', '외국인', '고령자', '외국인비율', '고령자비율']]
```
